Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. ## Предобработка данных

### Стиль заголовков

Посмотрим на названия наших столбцов.

```python
# перечень названий столбцов таблицы df
df.columns
```

Переименуем некоторые столбцы в snake_case и устраним заглавные буквы.

```python
# переименование столбцов
df = df.rename(columns={
    'cityCenters_nearest': 'city_сenters_nearest',
    'parks_around3000': 'parks_around_3000_meters',
    'ponds_around3000': 'ponds_around_3000_meters',
    'studio': 'is_studio',
    'open_plan': 'is_open_plan'
})
```

Проверим, что всё у нас получилось.

```python
# проверка результатов - перечень названий столбцов
df.columns
```

### Обработка пропусков

Сначала посчитаем, сколько в таблице пропущенных значений.

```python
# подсчёт пропусков
df.isna().sum()
```

Для наглядности посчитаем процентное соотношение пропусков в столбцах и подсветим.

```python
# выводим долю пропусков по столбцам
pd.DataFrame(round(
    df.isna().mean()*100, 2)).style.background_gradient('Oranges')
```

Сначала логично будет заполнить столбцы `is_apartment`, `balcony`, `parks_around_3000_meters` и `ponds_around_3000_meters`. В колонке `balcony` у нас указано количество балконов. Предположим, что пропуски означают их отсутствие, и заполним их явным обозначением `0`. `is_apartment` у нас указывает на то, является объект недвижимости апартаментами или нет, поэтому пропуски можно принять за отсутствие данного признака и заполнить их явным обозначением `0`. Аналогично для столбцов `parks_around_3000_meters` и `ponds_around_3000_meters`. Отсутствующие значения примем за нулевое количество объектов в радиусе 3000 метров, заменим на `0`. Столбец `days_exposition` может быть не заполнен, если объявление активно до сих пор, трогать не будем, чтобы не исказить результаты.

```python
# заполняем пропуски
df = df.fillna(value={
    'is_apartment': 0,
    'balcony': 0,
    'parks_around_3000_meters': 0,
    'ponds_around_3000_meters': 0}
)
```

Столбцы `floors_total` и `locality_name` содержат очень малое количество недостающих данных, поэтому удалим эти строки.

```python
df.dropna(subset=['floors_total', 'locality_name'], inplace=True)
```

Столбец `ceiling_height` отображает высоту потолков. Можно предположить, что пользователи по собственному желанию не стали его заполнять, так как не знают точно, какой высоты у них потолок. У нас пропушено почти 40% значений, но мы оставим данные без изменений, потому что заполнение такого количества данных одним характерным значением может привести к ошибкам.

Столбцы `airports_nearest`, `city_сenters_nearest`, `parks_nearest`, `ponds_nearest` относятся к данным, которые не заполняются самим пользователем, а вносятся автоматически. Трогать их не будем, запишем в Баг-репорт. Cтолбцы `living_area` и `kitchen_area` нам для исследования не пригодятся, поэтому не будем их трогать.

**Вывод**

Нам удалось заполнить пропуски в столбцах, где это вытекало из логики. Другие пропуски стоит отнести к ошибке при сборе данных.

### Замена типа данных

Столбцы `floors_total`, `balcony`, `parks_around_3000_meters`, `ponds_around_3000_meters` и `days_exposition` показывают количественные переменные, поэтому заменим тип данных на `int` вместо `float`, предварительно округлив значения.

```python
# преобразуем нужные столбцы
df['floors_total'] = df['floors_total'].round().astype(pd.Int64Dtype())
df['balcony'] = df['balcony'].round().astype(pd.Int64Dtype())
df['parks_around_3000_meters'] = df['parks_around_3000_meters'].round().astype(pd.Int64Dtype())
df['ponds_around_3000_meters'] = df['ponds_around_3000_meters'].round().astype(pd.Int64Dtype())
df['days_exposition'] = df['days_exposition'].round().astype(pd.Int64Dtype())
```

Столбец `is_apartment` по документации должен иметь булев тип.

```python
# преобразуем данный столбец
df['is_apartment'] = df['is_apartment'].astype('bool')
```

Столбец `first_day_exposition` содержит дату размещения объявления, но имеет тип данных `object`. Преобразуем его, оставим только саму дату, без времени.

```python
# преобразуем данный столбец
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%d')
```

Проверим, что всё удалось.

```python
df.info()
```

**Вывод**

Произвели необходимые преобразования типов данных в соответствии с документацией.

### Обработка дубликатов

Для начала приведём столбец `locality_name` с типом `object` к нижнему регистру.

```python
# применим метод lower()
df['locality_name'] = df['locality_name'].str.lower()
```

Проверим, есть ли дубликаты в нашей таблице.

```python
# посчитаем сумму дубликатов
print("Кол-во дубликатов: {}".format(df.duplicated().sum()))
```

**Вывод**

Проверили наличие дубликатов в данных. Не нашлось.

## Расчёты и добавление результатов в таблицу

### Расчёт цены за квадратный метр

Для расчёта цены за квадратный метр поделим значения столбца `last_price` на значения столбца `total_area`.

```python
# создаём новый столбец
df['per_sqare_meter_price'] = df['last_price'] / df['total_area']
# округлим дробные значения до двух знаков после запятой
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
# проверим, что у нас получилось
df.sample(10)
```

   ### Расчёт дня недели, месяца и года публикации объявления

Добавим ещё три необходимых столбца, исаользуя данные столбца `first_day_exposition`.

```python
# добавляем столбец weekday
df['weekday'] = df['first_day_exposition'].dt.weekday
# добавляем столбец month
df['month'] = df['first_day_exposition'].dt.month
# добавляем столбец year
df['year'] = df['first_day_exposition'].dt.year
```

```python
# взглянем на результат
df.sample(10)
```

Понедельник - первый день недели, обозначен под номером 0, а воскресенье - под номером 6. Месяцы обозначены цифрами от 1 до 12, годы обозначены номером года.

### Категоризация по этажу

Необходимо разделить записи по этажности: *первый*, *последний* и *другой*. Для этого напишем функцию и применим её при создании нового столбца.

```python
# функция категоризации по этажам
def floor_categorize(floor, last_floor):
    if floor == 1:
        return 'первый'
    elif floor == last_floor:
        return 'последний'
    else:
        return 'другой'
# создание нового столбца
df['floor_category'] = df.apply(lambda x: floor_categorize(x.floor, x.floors_total), axis=1)
```

```python
# посмотрим, что получилось
df.sample(10)
```

### Соотношение площадей внутри квартиры

Добавим очередные два столбца, один из которых будет содержать соотношение жилой площади к общей площади квартиры, другой будет содержать соотношение площади кухни к общей площади квартиры.

```python
# создаём новый столбец
df['living_to_total'] = df['living_area'] / df['total_area']
# создаём новый столбец
df['kitchen_to_total'] = df['kitchen_area'] / df['total_area']
```

```python
# проверим, что всё получилось
df.sample(10)
```

**Вывод**

Совершили необходимые расчёты и добавили их в нашу таблицу.

### Редкие и выбивающиеся значения

Создадим функцию подбора количества корзин.

```python
# функция подбора количества корзин
def bins_quantity(df, column):
    if df[column].max() > 1000:
        return 200
    elif df[column].max() > 500:
        return 100
    elif df[column].max() > 100:
        return 50
    elif df[column].max() > 50:
        return 20
    else:
        return 10
```

Очистим интересующие нас столбцы от экстремальных значений. Сначала выведем ящик с усами и гистограмму, 5 и 95 процентили. Создадим из них список и передадим его в цикл, который применит ко всем элементам нашу функцию. Затем очистим таблицу от значений ниже 5 процентиля и выше 95 процентиля.

```python
# список фильтруемых столбцов
columns_list = ['total_area', 'last_price', 'rooms', 'ceiling_height']
# функция вывода графиков    
def charts(df, column):    
    filtered_result = df.loc[(df[column] >= df[column].quantile(.01))&(
    df[column] <= df[column].quantile(.99)), column]
    plt.figure(figsize=(12, 5))
    plt.boxplot(filtered_result,
                notch=True,
                sym="o",
                vert=False,
                patch_artist=True)
    plt.ylabel('Boxplot ' + column)
    plt.tick_params(axis='both',
                    which='major',
                    direction='inout',
                    color='m',
                    )
    plt.grid()
    plt.show()
    plt.figure(figsize=(12, 5))
    hist = filtered_result.plot(
                kind='hist',
                bins=bins_quantity(df, column),
                alpha=0.5,
                ec='black',
                color='green',
                )
    plt.axvline(x=filtered_result.median(),
                linewidth=3,
                color='red',
                linestyle='--')
    plt.axvline(x=filtered_result.quantile(.01),
                linewidth=3,
                color='y',
                linestyle='--')
    plt.axvline(x=filtered_result.quantile(.99),
                linewidth=3,
                color='y',
                linestyle='--')
    plt.title('Распределение по столбцу ' + column)
    plt.xlabel('Значение')
    plt.ylabel('Частота')
    plt.grid()
    plt.show()
    print("Медиана: {0:.2f}".format(filtered_result.median()))
    print("1 процентиль: {0:.2f}".format(filtered_result.quantile(.01)))
    print("99 процентиль: {0:.2f}".format(filtered_result.quantile(.99)))
    print('--------------------------------------------------------------')
    
# цикл для фильтрации и вывода графиков
for column in columns_list:
    charts(df, column)
```

### Какие факторы больше всего влияют на стоимость квартиры?

Нам необходимо изучить, зависит ли цена от площади, числа комнат, удалённости от центра, зависимость цены от этажа и адты размещения объявления.

Создадим функцию, которая будет строить нам графики.

```python
# функция для построения графиков
def plot_per_sqare_meter_price(df, column, kind_name, style_name,
                               alpha_amount, hist_name, xlabel_name):
    (
        df.pivot_table(
            index=column,
            values='per_sqare_meter_price',
            aggfunc=['median'])
        .plot(kind=kind_name,
              style=style_name,
              y='median',
              grid=True,
              figsize=(12, 5),
              alpha=alpha_amount,
              title=hist_name)
    )
    plt.xlabel(xlabel_name)
    plt.ylabel('Цена за кв.м.')
    plt.show()
    try:
        print("Коэффициент корреляции: {0:.2f}".format(
            df[column].corr(df['per_sqare_meter_price'])))
    except:
        return
```

Далее создадим список значений, которые будем передавать в функцию с помощью цикла.
