Теперь объединим наши данные по столбцу `puNumber` методом `merge`, предварительно приведём столбец к одинаковому типу в обоих датафреймах. В `df_movies` у нас тип `object`, а в df_shows у нас тип `int`. Изменим тип столбца `puNumber` датафрейма `df_movies` на `int`, обработав его методом `to_numeric` с атрибутом `errors` равному `coerce`, чтобы принудительно возможные косяки в данных заменить на `NaN`. Необходимо сохранить все данные из датафрейма `df_movies`, поэтому при объединении укажем параметр `how` со значением `left`.

````python
# снчала применим метод to_numeric
df_movies['puNumber'] = pd.to_numeric(df_movies['puNumber'], errors='coerce')
# объединяем таблицы
df = df_movies.merge(df_shows, on='puNumber', how='left')
# проверяем результат
df.sample(10)
````

Столбец `distribution_certificate_number` содержит номер прокатного удостоверения, поэтому задаим ему тип данных `int`, так как при типе `object` длинное число будет неудобно отображаться. Аналогично для столбца `box_office`. В них есть пропуски, поэтому заменим их на `0`, чтобы они не мешали преобразованию. Пропущенные значения столбца `ratings` также заменим на `0` и преобразуем в `float`. Столбец `release_date` необходимо преобразовать в тип `datetime`. Оставшиеся столбцы соответствуют типу содержимого в них.

````python
# уникальные знчения столбца ratings
df['ratings'].unique()
````

Видим, что в столбец рейтинга попали значения рейтинга ожиданий. Нужно почистить. Создадим функцию, которая посимвольно разобьёт ячейку и положит все значения в список, если в списке будет присутствовать знак `%`, то заменим эту ячейку нулевым значением. Применим данную функцию к столбцу `ratings`.

````python
# функция разделения символов в ячейке
def clean_cell(cell):
    result = list(cell)
    if '%' in result:
        return 0
    else:
        return cell
# заполним пропуски
df['ratings'] = df['ratings'].fillna('%').astype('object')
# применим функцию
df['ratings'] = df['ratings'].apply(clean_cell)
````

````python
pd.to_numeric(df['ratings'], errors='coerce').unique()
````

Проверим результат.

````python
# уникальные знчения столбца ratings
df['ratings'].unique()
````

````python
# преобразуем столбцы distribution_certificate_number, box_office и ratings
df['distribution_certificate_number'] = df[
    'distribution_certificate_number'].fillna(0).astype('int')
df['box_office'] = df['box_office'].fillna(0).astype('int')
df['ratings'] = df['ratings'].astype('float')
# преобразуем столбец release_date, оставив только дату без времени
df['release_date'] = pd.to_datetime(df[
    'release_date']).apply(lambda x: x.strftime('%Y-%m-%d'))
df['release_date'] = pd.to_datetime(df['release_date'])
````

Взглянем на наши количественные столбцы.

````python
# зададим округление до двух знаков после запятой
pd.options.display.float_format = '{:.2f}'.format
# взглянем на количественные столбцы методом describe()
df[['budget', 'refundable_support', 'nonrefundable_support',
    'box_office']].describe()
````

По документации столбец `budget` включает в себя полный объём государственной поддержки, а потому должен содержать в себе сумму столбцов `refundable_support` и `nonrefundable_support`, как минимум, но может так же иметь не только государственный бюджет. Видим, что в нём есть нулевые значения, а такого быть не должно. Найдём значения, где общая сумма бюджета меньше, чем суммарная государственная поддержка и исправим на данную сумму.

````python
# проверим, какие строки удовлетворяют нашему условию
display(df.query('budget < (refundable_support + nonrefundable_support)'))
# посчитаем количество таких строк
print("Кол-во строк: {}".format(df.query(
   'budget < (refundable_support + nonrefundable_support)')['budget'].count()))
````

Видим, что все 16 строк содержат нулевой бюджет. Заменим эти значения на суммарную государственную поддержку.

````python
# замена аномальных значений
df.loc[df['budget'] < (df['refundable_support'] + df['nonrefundable_support']),
       'budget'] = df['refundable_support'] + df['nonrefundable_support']
````
