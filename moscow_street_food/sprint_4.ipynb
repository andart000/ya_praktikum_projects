## Анализ данных

### Исследуем виды объектов общепита

Посчитаем, какие вообще у нас виды есть в базе и построим горизонтальную гистограмму.

```python
# подсчёт количества уникальных значений в столбце
df['object_type'].value_counts()
```

Изменим значения в столбце, чтобы они были более читаемы для нас.

```python
# используем функцию replace
df = df.replace({'object_type': {'предприятие быстрого обслуживания': 'фастфуд',
                                 'магазин (отдел кулинарии)': 'кулинария'}})
```

```python
# зададим размер графиков
plt.rcParams['figure.figsize'] = (13,8)
# создаим функцию для построения горизонтальных гистограмм по нашим данным
def histogram(df, type, x_axis, y_axis, hue_name, xlimit, title_name):
    """Функция для построения наших графиков. Содержит несолько типов графиков,
    зависит от аргумента 'type':
    - countplot - гистограмма, которая принимает на вход один столбец, с отображением доли в %,
    считает количество значений в столбце;
    - countplot_with_hue - гисторамма, которая принимает на вход два столбца, без отображения долей,
    считает количество значений в столбцах;
    - barplot - гисторамма, которая принимает на вход два столбца, без отображения долей,
    выводит значения столбцов"""
    if type == 'countplot':
        horizontal_plot = sns.countplot(data=df,
                                        x=x_axis,
                                        y=y_axis, 
                                        orient='h',
                                        hue=hue_name,
                                        order=df[y_axis].value_counts().index)
        for i in horizontal_plot.patches:
            horizontal_plot.annotate(i.get_width(),
                                     xy=(i.get_width(),
                                     i.get_y()+i.get_height()/2),
                                     xytext=(10, 0), 
                                     textcoords='offset points', 
                                     ha="left", 
                                     va="center")
            horizontal_plot.annotate("%.f%%" % (i.get_width()/df.shape[0]*100),
                                     xy=(i.get_width(),
                                     i.get_y()+i.get_height()/2),
                                     xytext=(-5, 0), 
                                     textcoords='offset points', 
                                     ha="right", 
                                     va="center")
        sns.set_style('whitegrid')
        sns.set(font_scale=1.2)
        plt.xlim(0, xlimit)
        plt.title(title_name)
        plt.xlabel('Количество')
        plt.ylabel('Наименование')
        plt.show();
    elif type == 'countplot_with_hue':
        horizontal_plot = sns.countplot(data=df,
                                        y=y_axis,
                                        x=x_axis,
                                        orient='h',
                                        hue=hue_name,
                                        order=df[y_axis].value_counts().index)
        for i in horizontal_plot.patches:
            horizontal_plot.annotate(i.get_width(),
                                     xy=(i.get_width(),
                                     i.get_y()+i.get_height()/2),
                                     xytext=(10, 0), 
                                     textcoords='offset points', 
                                     ha="left", 
                                     va="center")
        sns.set_style('whitegrid')
        sns.set(font_scale=1.2)
        plt.xlim(0, xlimit)
        plt.title(title_name)
        plt.xlabel('Количество')
        plt.ylabel('Наименование')
        plt.show();
    elif type == 'barplot':
        horizontal_plot = sns.barplot(data=df, 
                                      y=y_axis,
                                      x=x_axis,
                                      hue=hue_name,
                                      orient='h',
                                      ci=None)                             
        for i in horizontal_plot.patches:
            horizontal_plot.annotate("%.f" % i.get_width(),
                                     xy=(i.get_width(),
                                     i.get_y()+i.get_height()/2),
                                     xytext=(10, 0), 
                                     textcoords='offset points', 
                                     ha="left", 
                                     va="center")
        sns.set_style('whitegrid')
        sns.set(font_scale=1.2)
        plt.xlim(0, xlimit)
        plt.title(title_name)
        plt.xlabel('Количество')
        plt.ylabel('Наименование')
        plt.show();
```

```python
# построим гистограмму по типам заведений
histogram(df,
          'countplot',
          None,
          'object_type',
          None,
          6500,
          'Виды объектов общепита')
```

### Создадим отдельный столбец с названием улицы

Выделим в отдельный столбец `street` информацию об улице из столбца `address`. В Зеленограде не используются названия улиц, обозначим поле `нет улицы`.

```python
# саздадим новый столбец
df = df.assign(street=lambda x: x['address'].str.split(',')
               .apply(lambda x: x[1] if 'город' not in x[1] and
               'поселение' not in x[1] else (
                   x[2] if 'корпус' not in x[2] else 'нет улицы')))
```

Проверим, что всё получилось.

```python
# выведем 10 случайных строк
df.sample(10)
```

### Рассмотрим топ-10 улиц с наибольшим количеством объектов

Создадим отдельный датафрейм, где посчитаем количество заведений на каждой из улиц. Отсортируем по убыванию и оставим только топ-10. Значения `нет улицы` исключим из расчётов.

```python
# датафрейм с топ-10 улиц по количеству объектов
df_streets_top = (
    df.loc[df['street'] != 'нет улицы'].groupby(['street'])
    .agg(amount=('id','count'))
    .sort_values(by='amount', ascending=False)
    .reset_index()
    .loc[:10]
)
```

Посмотрим, что получилось.

```python
# выводим результат
df_streets_top
```

```python
# построим гистограмму топ-10 улиц с наибольшим количеством объектов
histogram(df_streets_top,
          'barplot',
          'amount',
          'street',
          None,
          250,
          'Количество объектов на улице')
```

Отфильруем датафрейм, оставив в нём только заведения, расположенные на топ-10 улицах.

```python
# создадим список с наиболее популярными улицами
df_streets_top_list = df_streets_top['street'].to_list()
# выведем результат
df_streets_top_list
```

```python
# фильтруем датафрейм
df_filtered = df[df['street'].isin(df_streets_top_list)]
# выводим результат
df_filtered
```

Далее сохраним это всё дело в файл, чтобы лишний раз не обращаться к API. Позднее, с помощью этих координат мы сможем запросить название муниципального района.

```python
# создадим функцию, которая получит для нас координаты через API Яндекс.Карт
def get_coordinates(address):
    base_url = "https://geocode-maps.yandex.ru/1.x"
    response = requests.get(base_url, params={
        "geocode": address,
        "apikey": '42096618-c47f-45f8-bca4-2c6eb2cbd672',
        "format": "json",
    })
    response.raise_for_status()
    found_places = response.json()['response']['GeoObjectCollection']['featureMember']

    if not found_places:
        return None

    most_relevant = found_places[0]
    coordinates = most_relevant['GeoObject']['Point']['pos'].split(' ')
    return coordinates
```
